Plot (using Matplotlib) a spot diagram using <font color='blue'>`spiralSpot()`</font> function

```python
x,y,_,_ = l.spiralSpot(0.0,0.4,1,10,1000)
```

```python
sp = scatter(x, y, s=5, c='red', linewidth=0.35)
grid()
ax = gca()
ax.set_aspect('equal')
```

```python
if l.zPushLensPermission():
    l.zPushLens(updateFlag=1)
```
