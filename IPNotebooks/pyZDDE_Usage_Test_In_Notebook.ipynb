# <font color='DarkBrown'>Using PyZDDE with IPython webnotebook</font>

Using the IPython webnotebook with Zemax using the PyZDDE toolbox is an attractive application because it can provide a way for quick documentation of lens design process as one progress through the design, including embedding intermittent figures, prescription files, thought process, etc. It can also be used in an educational setting.

There are mainly 2 ways of embedding the graphic windows from within Zemax into the Webnotebook (or IPython QtConsole) -- using the commands `ipCaptureWindow` and `ipCaptureWindow2`. `ipCaptureWindow` basically exports JPG images from Zemax and embeds them into the IPython shell. The quality of the embedded JPG image is limited by the JPG export quality from Zemax. If better quality graphics is desired, then use `ipCaptureWindow2` requests Zemax to export the grahics as Metafile (WMF/EMF), resizes, converts to PNG, and then embeds them into the IPython shell. 

##<font color='Brown'>Using `ipCaptureWindow`:</font>

The function `ipCaptureWindow` is useful for quickly capturing a graphic window, and embedding into an IPython Webnotebook or QtConsole. 

In order to use this function, please copy the ZPL macros from PyZDDE\ZPLMacros to the macro directory where Zemax is expecting the ZPL macros to be (i.e. the folder set in Zemax->Preference->Folders->ZPL).
